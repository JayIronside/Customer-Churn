# Importing data for use

```python
import pandas as pd
```

```python
data = pd.read_csv('Customer-churn.csv')
```

# Feature Selection

```python
feature = data[['gender', 'PhoneService', 'PaperlessBilling', 'StreamingTV', 'MonthlyCharges', 'TotalCharges' ]]
```

```python
feature
```

```python
target = data[['Churn']]
```

```python
target
```

# Categorical values to Numerical values

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
feature["gender"] = le.fit_transform(feature["gender"])
```

```python
feature['PhoneService'] = le.fit_transform(feature['PhoneService'])
```

```python
feature['PaperlessBilling'] = le.fit_transform(feature['PaperlessBilling'])
```

```python
feature['StreamingTV'] = le.fit_transform(feature['StreamingTV'])
```

```python
feature
```

```python
target = le.fit_transform(data['Churn'])
```

```python
target
```

# Splitting data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
feature_train, feature_test, target_train, target_test = train_test_split(feature, target, test_size = 0.3)
```

```python
feature_train
```

```python
target_train
```

```python
feature_train["TotalCharges"] = le.fit_transform(feature_train["TotalCharges"])
```

```python
feature_train
```

```python
feature_test["TotalCharges"] = le.fit_transform(feature_test["TotalCharges"])
```

# Building the model

```python
from sklearn.svm import SVC
classifier = SVC()
model = classifier.fit(feature_train, target_train)
prediction = model.predict(feature_test)
```

```python
prediction
```

```python
dictionary = {"Actual Value": target_test, "Predictions": prediction}
dictionary
```

```python
# Displaying the dictionary as a table
GraphicalRep = pd.DataFrame(dictionary)
```

```python
GraphicalRep.tail(30)
```

# Evaluate the model

```python
from sklearn.metrics import accuracy_score
accuracy_score(target_test, prediction)
```

```python
# using confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(target_test, prediction))
```
